# Total unpaid per instalment_plan_id

```python
# Create unpaid variables per instalment_plan_id
unpaid_per_instalment_plan_id = instalments_df[instalments_df['status'] == 'unpaid'].groupby('instalment_plan_id').sum()[['penalty_fee', 'amount', 'total']].reset_index().rename(columns={'penalty_fee':'total_penalty_fee', 'amount':'total_unpaid_amount', 'total':'total_unpaid_total'})
```

```python
# Create unpaid variables per instalment_plan_id
amount_per_instalment_plan_id = instalments_df.groupby('instalment_plan_id').sum()[['penalty_fee', 'amount', 'total']].reset_index().rename(columns={'penalty_fee':'total_penalty_fee', 'amount':'total_unpaid_amount', 'total':'total_unpaid_total'})
amount_per_instalment_plan_id.head()
```

```python
# Create days diff: dyas between scheduled and completed; days between scheduled and today
instalments_df['days_scheduled_completed'] = (pd.to_datetime(instalments_df['completed']) - pd.to_datetime(instalments_df['scheduled'])).dt.days
instalments_df['days_since_scheduled'] = (datetime.datetime.today() - pd.to_datetime(instalments_df['scheduled'])).dt.days
```
